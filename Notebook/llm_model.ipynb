## Section 2: Load HuggingFace model

```
from transformers import pipeline

medical_ai = pipeline(
    "text-generation",
    model="google/medgemma-1.5-4b-it"
)
```

##Section 3: Explanation generator

```
def generate_medical_report(result):

    prompt = f"""
A chest X-ray AI model predicted the following result.

Disease: {result['disease']}
Confidence: {result['confidence']}

Class probabilities:
{result['probabilities']}

Explain this diagnosis like a doctor.

Include:
1. What the disease is
2. Causes
3. Recommended tests
4. Treatment
5. Precautions
"""

    response = medical_ai(prompt, max_length=200)

    return response[0]["generated_text"]
```

## Section 4: Example test

```
sample_result = {
    "disease": "COVID",
    "confidence": 0.91,
    "probabilities": {
        "COVID": 0.91,
        "Normal": 0.04,
        "Lung_Opacity": 0.03,
        "Viral Pneumonia": 0.02
    }
}

report = generate_medical_report(sample_result)

print(report)
```
